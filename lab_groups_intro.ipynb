Проверим известные нам свойства знакопеременной группы $A_4$

```python
G = AlternatingGroup(4)
```

```python
G.order() # порядок группы
```

```python
for g in G:
    print(g)
```

Далее изучим (а точнее вспомним) подгруппы группы $A_4$. Здесь работает команда `subgroups`.

```python
subgrs = G.subgroups()
```

```python
for H in subgrs:
    print(H)
```
